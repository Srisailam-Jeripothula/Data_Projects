# **Aerofit Business Case Study**

## **Introduction**
Aerofit is a leading brand in the field of fitness equipment, providing a diverse range of products including treadmills, exercise bikes, gym equipment, and fitness accessories to cater to the needs of all categories of people. The **Market Research Team** at Aerofit is keen on identifying the characteristics of the target audience for each type of treadmill they offer. To provide better recommendations to new customers and optimize marketing efforts, it is essential to perform **descriptive analysis** to create a customer profile for each product by using tables and charts, and to compute **marginal and conditional probabilities** along with extracting meaningful insights.  

---

## **Dataset Overview**
The dataset contains information about customer characteristics and the type of treadmill they purchased. The key attributes include:  

- **Product:** Product Purchased – KP281, KP481, or KP781  
- **Age:** Age in years  
- **Gender:** Male/Female  
- **Education:** Years of education  
- **MaritalStatus:** Single or partnered  
- **Usage:** Average number of times the customer plans to use the treadmill each week  
- **Income:** Annual income in dollars ($)  
- **Fitness:** Self-rated fitness on a 1-to-5 scale (1 = poor shape, 5 = excellent shape)  
- **Miles:** Average number of miles the customer expects to walk/run each week  

The dataset provides valuable insights into customer characteristics and their purchasing behavior, which will help us understand the customer profiles associated with different treadmill products.

---

## **Product Portfolio**
Aerofit offers three types of treadmills, each designed for a different segment of users:  

- **KP281:** An entry-level treadmill priced at 1500
- **KP481:** A mid-level treadmill for runners priced at 1750  
- **KP781:** An advanced treadmill with premium features priced at 2,500  

---

## **Objective and Scope of Analysis**
The objective of this analysis is to create detailed **customer profiles** for each Aerofit treadmill product by analyzing customer data. The analysis will provide insights to target customers more effectively and design personalized marketing campaigns.  

The scope of this analysis includes:  

- **Customer Profile Creation:**  
    - Analyzing customer demographics to determine which age groups, income levels, and fitness levels prefer each type of treadmill.  
    - Understanding which gender prefers which treadmill model.  
    - Identifying patterns in treadmill usage and miles expected to be walked or run by customers.  

- **Descriptive Analytics:**  
    - Creating visualizations to identify trends and patterns among customers for each treadmill type.  
    - Understanding the relationship between customer characteristics and product preferences.  

- **Probability Analysis:**  
    - Computing **marginal probabilities** to assess the likelihood of customers purchasing each treadmill.  
    - Determining **conditional probabilities** to identify the probability of purchasing a specific treadmill based on customer characteristics such as gender, income, or fitness level.  

---

## **Customer Profiling and Insights**
To create a customer profile for each Aerofit treadmill product, we will analyze data to determine:  

- **What type of customers are buying which type of treadmill?**  
- **Which gender is more inclined to buy a particular treadmill?**  
- **Which income sector prefers which type of treadmill?**  
- **What age groups opt for entry-level, mid-level, and advanced treadmills?**  
- **How are these metrics related to each other?**  

By performing this analysis, we will gain a better understanding of customer preferences, which will guide marketing campaigns and assist in designing advertisements targeted at specific customer segments. These insights will enable Aerofit to enhance its market presence and provide better recommendations to new customers.

---

## **Conclusion**
By performing detailed analysis of the customer data, Aerofit will be able to identify key characteristics of its target audience and refine its marketing strategy. Understanding customer preferences for each type of treadmill will not only improve product recommendations but also boost customer satisfaction and increase revenue.

```python
columns=[ 'Age', 'Education', 'Usage','Fitness', 'Income', 'Miles']
clipped_df=df.copy()
for col in columns:
  p5,p95=np.percentile(df[col],[5,95])
  clipped_df[col]=np.clip(df[col],p5,p95)
clipped_df.head()
```

### **Insights from Clipping:**
**Reduced Influence of Outliers:**  
   - Clipping successfully handled outliers beyond the 5th and 95th percentiles.
   - Extreme values that could have skewed the analysis were effectively managed.

**Preserved Data Integrity:**  
   - The core data distribution remained intact, ensuring that meaningful trends and patterns were not distorted.
   - Only extreme values were clipped, minimizing data loss.

**Improved Visualization and Analysis:**  
   - The boxplots show a more compact distribution post-clipping, making the data better suited for statistical analysis.
   - Clipped data enhances the performance of machine learning models by reducing noise caused by extreme values.

**Consistency Across Multiple Variables:**  
   - The clipping process was applied to multiple columns (Age, Education, Usage, Fitness, Income, and Miles), ensuring consistent treatment of outliers across all relevant metrics.


```python
clipped_df.describe()
```

# **Before and After Clipping: Statistical Comparison**


## **Key Observations**
**Reduction in Extremes**  
   - The maximum values for **Age (50 → 43.05)**, **Miles (360 → 200)**, and **Income (1,04,581 → 90,948.25)** have been significantly reduced.
   - This suggests that extreme outliers were removed or adjusted.

**Effect on Mean & Standard Deviation**  
   - The **mean of Income** slightly decreased (**53577.78 → 53477.07**), indicating that high-income outliers were affecting the earlier average.
   - **Miles per week increased** after clipping (**82.56 → 101.09**), meaning some extreme low values were also adjusted.

**Median and Quartiles**  
   - The **median values remain mostly unchanged**, indicating that the core distribution was not heavily impacted.
   - The **25th and 75th percentiles** for all variables have slightly shifted, reflecting better data consistency.

**Fitness and Usage Stability**  
   - The **fitness and usage distributions remain largely the same**, meaning clipping had little impact on these features.

---

### **Conclusion**
- The clipping process effectively removed extreme outliers while preserving the overall structure of the data.
- The changes in mean and max values indicate that previously, a few extreme values were skewing the analysis.
- This improved dataset now provides **more reliable insights** for customer profiling and product recommendations.
